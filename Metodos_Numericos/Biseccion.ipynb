# Metodo de la Biseccion comparado con otros metodos para hallar raices

```python
import math
import pylab as pl

i1 = []
i2 = []
i3 = []
i4 = []
i11 = []
i22 = []
i33 = []
i44 = []
NewA = []
SecA = []
PunA = []
BisA = []
NewT = []
SecT = []
PunT = []
BisT = []
error = 0.00005
xl = 1
xu = 0

def funcion(xi):
    return ((math.exp(-(xi)))-xi)
    
def derivada(xi):
    return (-(math.exp(-(xi)))-1)

def Newton():
    xi = 1
    cont = 0
    errorA = 1
    print("i\t   xi\t\t   Ea\t\t   Et")
    while errorA >= error:
        sig = xi - (funcion(xi)/derivada(xi))
        errorA = abs(((sig - xi)/sig)*100)
        errorT = abs(((0.56714329 - sig)/0.56714329)*100)
        print("%d\t%f\t%f\t%f") % (cont,xi,errorA, errorT)
        NewA.append(errorA)
        NewT.append(errorT)
        i2.append(cont)
        i22.append(cont)
        xi = sig
        cont = cont + 1
        
def Secante():
    xi = 1
    ant = 0
    cont = 0
    errorP = 1
    print("i\t   xi\t\t   Ea\t\t   Et")
    while errorP >= error:
        sig = xi - ((funcion(xi))*(ant - xi))/(funcion(ant)-funcion(xi))
        errorA = abs(((sig - xi)/sig)*100)
        errorT = abs(((0.56714329 - sig)/0.56714329)*100)
        print("%d\t%f\t%f\t%f") % (cont,xi,errorA, errorT)
        SecA.append(errorA)
        SecT.append(errorT)
        i3.append(cont)
        i33.append(cont)
        ant = xi
        xi = sig
        if cont < 6:
            cont = cont + 1
        else:
            break

def PuntoFijo():
    xi = 1
    cont = 0
    errorA = 1
    print("i\t   xi\t\t   Ea\t\t   Et")
    while errorA >= error:
        sig = math.exp(-xi)
        errorA = abs(((sig - xi)/sig)*100)
        errorT = abs(((0.56714329 - sig)/0.56714329)*100)
        print("%d\t%f\t%f\t%f") % (cont,xi,errorA, errorT)
        i1.append(cont)
        i11.append(cont)
        PunA.append(errorA)
        PunT.append(errorT)
        xi = sig
        cont = cont + 1
        
def Biseccion():
    xl = 1
    xu = 0
    xi = 1
    cont = 0
    errorA = 1
    print("i\t   xi\t\t   Ea\t\t   Et")
    while errorA >= error:
        xr = (xl + xu)*0.5
        if (funcion(xl)*funcion(xr) < 0):
            xu = xr
            errorA = abs(((xr - xi)/xr)*100)
            errorT = abs(((0.56714329 - xr)/0.56714329)*100)
            print("%d\t%f\t%f\t%f") % (cont,xr,errorA, errorT)
            BisA.append(errorA)
            BisT.append(errorT)
        elif (funcion(xl)*funcion(xr) > 0):
            xl = xr 
            errorA = abs(((xr - xi)/xr)*100)
            errorT = abs(((0.56714329 - xr)/0.56714329)*100)
            print("%d\t%f\t%f\t%f") % (cont,xr,errorA, errorT)
            BisA.append(errorA)
            BisT.append(errorT)
        elif (funcion(xl)*funcion(xr) == 0):
            print ("La raiz de la funcion es: %f") % xr
        i4.append(cont)
        i44.append(cont)
        xi = xr
        cont = cont + 1

print("\nMetodo del Punto Fijo\n")
PuntoFijo()
print("\nMetodo de Newton - Raphson\n")
Newton()
print("\nMetodo de la Secante\n")
Secante()
print("\nMetodo de la Biseccion\n")
Biseccion()

pl.grid(True)
pl.figure(1)
pl.subplot(211)
pl.plot(i11, PunT, 'g', label = "Punto Fijo")
pl.plot(i22, NewT, 'k', label = "Newton")
pl.plot(i33, SecT, 'b', label = "Secante")
pl.plot(i44, BisT, 'r', label = "Biseccion")
pl.legend(loc='upper right')
pl.xlabel("Iteracion")
pl.ylabel("Error Relativo")
pl.title("Metodos para Hallar Raices")

pl.subplot(212)
pl.plot(i1, PunA, 'g', label = "Punto Fijo")
pl.plot(i2, NewA, 'k', label = "Newton")
pl.plot(i3, SecA, 'b', label = "Secante")
pl.plot(i4, BisA, 'r', label = "Biseccion")
pl.xlim(0.0, 10)
pl.ylim(0.0, 200)
pl.legend(loc='upper right')
pl.xlabel("Iteracion")
pl.ylabel("Error Aproximado")

pl.show()
```
