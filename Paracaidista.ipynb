# Solución analítica del problema del paracaidista que cae

```python
import math
import numpy as np
import pylab as pl

masa = 68.1
resistencia = 12.5
gravedad = 9.8
error = 0.0005
cont = 0
velocidad_actual = 0
velocidad_aprox = 0
velocidad_exacta = 0
errores = 1
contador = []
velocidad_ex = []
velocidad_ap = []

for x in xrange(100):
    if errores >= error:
        velocidad_exacta = ((gravedad*masa)/resistencia)*(1-math.exp(-(resistencia/masa)*cont))
        velocidad_aprox = velocidad_actual+(gravedad-((resistencia/masa)*velocidad_actual))*((cont+1)-cont)
        velocidad_actual = velocidad_aprox
        contador.append(cont)
        cont = cont + 1
        velocidad_ex.append(velocidad_exacta)
        velocidad_ap.append(velocidad_aprox)
        if cont > 1:
            errores = (velocidad_aprox-velocidad_exacta)/velocidad_exacta
        print ("%d ------------------") % cont
        print ("La velocidad exacta es: %f") % velocidad_exacta
        print ("La velocidad aproximada es: %f") % velocidad_aprox
        print ("El error es: %f") % errores
        
ex = pl.plot(contador, velocidad_ex, "r")
ap = pl.plot(contador, velocidad_ap, "g")
pl.title("Velocidades de caida")
pl.xlabel("Tiempo")
pl.ylabel("Velocidad")
pl.xlim(0.0, 50)
pl.ylim(0.0, 60)
pl.show()
```
