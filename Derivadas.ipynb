# Derivadas de 3 ecuaciones. Calculo de derivada real, por delante, por atr√°s y centrada, y una grafica que las represente.

```python
import math
import numpy as np
import matplotlib.pyplot as pl

resultado = 0
primAdel = []
primAtras = []
primCent = []
segAdel = []
segAtras = []
segCent = []
terAdel = []
terAtras = []
terCent = []
primVerd= []
segVerd = []
terVerd = []
ejex = []
errorA = 0
errorR = 0
ecu1 = "(e^x)-1"
ecu2 = "sen^3(2x)+x^2"
ecu3 = "x^4+3x^2-cos(x)"

def primeraEcuacion(x):
    resultado = math.exp(x)-1
    return resultado

def segundaEcuacion(x):
    resultado = (math.sin(2*x)**3) + (x**2)
    return resultado

def terceraEcuacion(x):
    resultado = (x**4) + 3*(x**2) - math.cos(x)
    return resultado

def errorAbsoluto(primera, valor):
    return abs(primera - valor)
            
def errorRelativo(primera, valor):
    return (primera - valor)/ primera

def derivadas(h):
    x = -5.0
    while x <= 5:
        primera = math.exp(x)
        adelante1 = (primeraEcuacion(x+h)-primeraEcuacion(x))/h
        atras1 = (primeraEcuacion(x)-primeraEcuacion(x-h))/h
        centrada1 = (primeraEcuacion(x+h)-primeraEcuacion(x-h))/(2*h)
        abAdel1 = errorAbsoluto(primera, adelante1)
        abAtras1 = errorAbsoluto(primera, atras1)
        abCent1 = errorAbsoluto(primera, centrada1)
        relAdel1 = errorRelativo(primera, adelante1)
        relAtras1 = errorRelativo(primera, atras1)
        relCent1 = errorRelativo(primera, centrada1)
        porAdel1 = errorRelativo(primera, adelante1)*100
        porAtras1 = errorRelativo(primera, atras1)*100
        porCent1 = errorRelativo(primera, centrada1)*100
        
        primVerd.append(primera)
        primAdel.append(adelante1)
        primAtras.append(atras1)
        primCent.append(centrada1)
        
        segunda = (6*(math.sin(2*x)**2)*(math.cos(2*x)))+(2*x)
        adelante2 = (segundaEcuacion(x+h)-segundaEcuacion(x))/h
        atras2 = (segundaEcuacion(x)-segundaEcuacion(x-h))/h
        centrada2 = (segundaEcuacion(x+h)-segundaEcuacion(x-h))/(2*h)
        abAdel2 = errorAbsoluto(primera, adelante2)
        abAtras2 = errorAbsoluto(primera, atras2)
        abCent2 = errorAbsoluto(primera, centrada2)
        relAdel2 = errorRelativo(primera, adelante2)
        relAtras2 = errorRelativo(primera, atras2)
        relCent2 = errorRelativo(primera, centrada2)
        porAdel2 = errorRelativo(primera, adelante2)*100
        porAtras2 = errorRelativo(primera, atras2)*100
        porCent2 = errorRelativo(primera, centrada2)*100
        
        segVerd.append(segunda)
        segAdel.append(adelante2)
        segAtras.append(atras2)
        segCent.append(centrada2)
        
        tercera = 4*(x**3) + (6*x) + (math.sin(x))
        adelante3 = (terceraEcuacion(x+h)-terceraEcuacion(x))/h
        atras3 = (terceraEcuacion(x)-terceraEcuacion(x-h))/h
        centrada3 = (terceraEcuacion(x+h)-terceraEcuacion(x-h))/(2*h)
        abAdel3 = errorAbsoluto(primera, adelante3)
        abAtras3 = errorAbsoluto(primera, atras3)
        abCent3 = errorAbsoluto(primera, centrada3)
        relAdel3 = errorRelativo(primera, adelante3)
        relAtras3 = errorRelativo(primera, atras3)
        relCent3 = errorRelativo(primera, centrada3)
        porAdel3 = errorRelativo(primera, adelante3)*100
        porAtras3 = errorRelativo(primera, atras3)*100
        porCent3 = errorRelativo(primera, centrada3)*100    
        
        terVerd.append(tercera)
        terAdel.append(adelante3)
        terAtras.append(atras3)
        terCent.append(centrada3)
        
        ejex.append(x)
        x = x + h

    pl.plot(ejex, primVerd, label = "Valor Real")
    pl.plot(ejex, primAdel, label = "Adelante")
    pl.plot(ejex, primAtras, label = "Atras")
    pl.plot(ejex, primCent, label = "Centrada")
    pl.title("Primera Ecuacion: "+ecu1)
    pl.xlabel("Valores de X")
    pl.ylabel("Resultado")
    pl.legend(loc = "upper left") 
    pl.show()
    
    pl.plot(ejex, segVerd, label = "Valor Real")
    pl.plot(ejex, segAdel, label = "Adelante")
    pl.plot(ejex, segAtras, label = "Atras")
    pl.plot(ejex, segCent, label = "Centrada")
    pl.title("Segunda Ecuacion: "+ecu2)
    pl.xlabel("Valores de X")
    pl.ylabel("Resultado")
    pl.legend(loc = "upper left") 
    pl.show()
    
    pl.plot(ejex, terVerd, label = "Valor Real")
    pl.plot(ejex, terAdel, label = "Adelante")
    pl.plot(ejex, terAtras, label = "Atras")
    pl.plot(ejex, terCent, label = "Centrada")
    pl.title("Primera Ecuacion: "+ecu3)
    pl.xlabel("Valores de X")
    pl.ylabel("Resultado")
    pl.legend(loc = "upper left") 
    pl.show()
    
h = input("Ingrese el valor de h: ")
derivadas(h)
      
```
