# Ejercicio 17.1

```python
import math

Y = [8.8,9.5,9.8,9.4,10.0,9.4,10.1,9.2,11.3,9.4,10.0,10.4,7.9,10.4,9.8,9.8,9.5,8.9,8.8,10.6,10.1,9.5,9.6,10.2,8.9]
n = 25

def Mediay():
    y = 0.0
    aux = 0.0
    suma = 0.0
    for i in range(n):
        aux = Y[i]
        suma = suma + aux
    y = suma/n
    print ("a) La media es: %f") % y
    return y

def DEstandar():
    suma = 0.0
    exp = 0.0
    var = 0.0
    dEst = 0.0
    m = Mediay()
    CV = 0.0
    for i in range(n):
        aux = Y[i] - m
        exp = aux**2
        suma = suma + exp    
    var = suma/(n-1)
    dEst = abs(var**0.5)
    CV = (dEst/m)*100
    print ("b) La desviacion estandar es: %f") % dEst
    print ("c) La varianza es: %f") % var
    print ("d) El coeficiente de variacion es: %f") % CV
    intAnt = m-(2*dEst)
    intSig = m+(2*dEst)
    print ("e) El intervalo de confianza del 95 debe estar entre: %f y %f") % (intAnt,intSig)
    
DEstandar()
```

# Ejercicio 17.4

```python
import math
import pylab as pl

Y = [0,2,4,6,9,11,12,15,17,19]
X = [5.0,6.0,7.0,6.0,9.0,8.0,7.0,10.0,12.0,12.0]
n = 10
L = []
sumx = 0.0
sumy = 0.0
cuax = 0.0
cuay = 0.0

for i in range(n):
    sumx = sumx + X[i]
    cuax = cuax + (X[i]**2)
    
for j in range(n):
    sumy = sumy + Y[j]
    cuay = cuay + (Y[j]**2)  
    
def Suma():
    suma = 0.0
    aux = 0.0
    for i in range(n):
        aux = X[i]*Y[i]
        suma = suma + aux
    return suma

def Mediay():
    y = 0.0
    aux = 0.0
    suma = 0.0
    for i in range(n):
        aux = Y[i]
        suma = suma + aux
    y = suma/n
    return y

def Mediax():
    x = 0.0
    aux = 0.0
    suma = 0.0
    for i in range(n):
        aux = X[i]
        suma = suma + aux
    x = suma/n
    return x

def DEstandar():
    suma = 0.0
    exp = 0.0
    var = 0.0
    dEst = 0.0
    m = Mediay()
    for i in range(n):
        aux = Y[i] - m
        exp = aux**2
        suma = suma + exp    
    var = suma/(n-1)
    dEst = abs(var**0.5)
    
def Regresion():
    suma = 0.0
    Sr = 0.0
    suma = Suma()
    a1 = ((n*suma) - (sumx*sumy))/((n*cuax) - (sumx**2))
    a0 = Mediay() - (a1*Mediax())
    for i in range(n):
        y = a0 + a1*X[i]
        L.append(y)
    for j in range(n):
        aux = (Y[j]-a0-a1*X[j])**2
        Sr = Sr + aux
    aux = Sr/(n-2)
    error = aux**0.5
    print ("El error estandar de estimacion XvsY es: %f") % error
    
def Correlacion():
    
    sumy = 0.0
    sumx = 0.0
    cuay = 0.0
    cuax = 0.0
    r = 0.0
    
    for k in range(n):
        sumy = sumy + Y[k]
        sumx = sumx + X[k]
        cuax = cuax + (X[k]**2) 
        cuay = cuay + (Y[k]**2)
    
    xy = 0.0
    aux = 0.0
    for i in range(n):
        aux = Y[i]*X[i]
        xy = xy + aux
        
    r = ((n*xy) - (sumy*sumx))/((np.sqrt((n*cuax) - (sumx**2)))*(np.sqrt((n*cuay) - (sumy**2))))
    print("El Coeficiente de Correlacion XvsY es: %f") % r
    
DEstandar()
Regresion()
Correlacion()

pl.plot(X, Y, 'o', label = "Valores")
pl.plot(X, L, label = "Regresion Lineal")
pl.legend(loc = "lower right")
pl.title("Ejercicio 17.4")
pl.xlabel("Valores X")
pl.ylabel("Valores Y")
pl.grid(True)
pl.show()

X = [0,2,4,6,9,11,12,15,17,19]
Y = [5.0,6.0,7.0,6.0,9.0,8.0,7.0,10.0,12.0,12.0]
n = 10
L = []
sumx = 0.0
sumy = 0.0
cuax = 0.0
cuay = 0.0

for i in range(n):
    sumx = sumx + X[i]
    cuax = cuax + (X[i]**2)
    
for j in range(n):
    sumy = sumy + Y[j]
    cuay = cuay + (Y[j]**2)  
    
def Suma():
    suma = 0.0
    aux = 0.0
    for i in range(n):
        aux = X[i]*Y[i]
        suma = suma + aux
    return suma

def Mediay():
    y = 0.0
    aux = 0.0
    suma = 0.0
    for i in range(n):
        aux = Y[i]
        suma = suma + aux
    y = suma/n
    return y

def Mediax():
    x = 0.0
    aux = 0.0
    suma = 0.0
    for i in range(n):
        aux = X[i]
        suma = suma + aux
    x = suma/n
    return x

def DEstandar():
    suma = 0.0
    exp = 0.0
    var = 0.0
    dEst = 0.0
    m = Mediay()
    for i in range(n):
        aux = Y[i] - m
        exp = aux**2
        suma = suma + exp    
    var = suma/(n-1)
    dEst = abs(var**0.5)
    
def Regresion():
    suma = 0.0
    Sr = 0.0
    suma = Suma()
    a1 = ((n*suma) - (sumx*sumy))/((n*cuax) - (sumx**2))
    a0 = Mediay() - (a1*Mediax())
    for i in range(n):
        y = a0 + a1*X[i]
        L.append(y)
    for j in range(n):
        aux = (Y[j]-a0-a1*X[j])**2
        Sr = Sr + aux
    aux = Sr/(n-2)
    error = aux**0.5
    print ("El error estandar de estimacion YvsX es: %f") % error
    
def Correlacion():
    
    sumy = 0.0
    sumx = 0.0
    cuay = 0.0
    cuax = 0.0
    r = 0.0
    
    for k in range(n):
        sumy = sumy + Y[k]
        sumx = sumx + X[k]
        cuax = cuax + (X[k]**2) 
        cuay = cuay + (Y[k]**2)
    
    xy = 0.0
    aux = 0.0
    for i in range(n):
        aux = Y[i]*X[i]
        xy = xy + aux
        
    r = ((n*xy) - (sumy*sumx))/((np.sqrt((n*cuax) - (sumx**2)))*(np.sqrt((n*cuay) - (sumy**2))))
    print("El Coeficiente de Correlacion YvsX es: %f") % r
    
DEstandar()
Regresion()
Correlacion()

pl.plot(X, Y, 'o', label = "Valores")
pl.plot(X, L, label = "Regresion Lineal")
pl.legend(loc = "lower right")
pl.title("Ejercicio 17.4")
pl.xlabel("Valores X")
pl.ylabel("Valores Y")
pl.grid(True)
pl.show()
```
